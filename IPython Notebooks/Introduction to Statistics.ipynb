```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Scatter plot
x = [1300, 1600, 1700, 1900, 2100, 2200]
y = [82000, 50000, 53000, 59000, 44000, 68000]
plt.plot(x, y)
plt.xlabel('Square foot')
plt.ylabel('Price')
plt.title('Housing prices')
```

```python
#Solve system of linear equation
x = [[1700, 2100], [1, 1]]
print x
y = [53000, 65000]
soln = np.linalg.solve(x, y)
print soln
np.allclose(np.dot(x, soln), y)
```

```python
# Bar Chart
x = [1300, 1400, 1600, 1900, 2100, 2300]
y = [88000, 73000, 94000, 86000, 112000, 98000]
x2 = [1100, 900, 1200, 3400, 1600]
y2 = [81000, 32000, 120000, 33000, 70000]
# x = np.random.randn(1000)
plt.bar(x, y, width=100, color='blue', alpha=0.5)
plt.bar(x2, y2, width=100, color='green', alpha=0.5)
```

```python
# Histogram
# Difference in Histogram and bar chart is, barcharts are 2D while histograms are 1D
# y label in histograms is the counts
x = [10, 20, 30, 35, 50, 55]
plt.hist(x, bins=10)
```
